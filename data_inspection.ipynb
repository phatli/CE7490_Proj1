```python
import os
import sys

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20, 3)
```

```python
root_path = os.path.join(os.path.dirname(os.getcwd()+"/"), 'data/azurefunctions-dataset2019/')
print(f"Data path: {root_path}")

files = os.listdir(root_path)
function_files = [f for f in files if f[0] == 'f']
invocation_files = [f for f in files if f[0] == 'i']
app_files = [f for f in files if f[0] == 'a']

print(f"Funtion files: \n {function_files}")
print(f"Invocation files: \n {invocation_files}")
print(f"App files: \n {app_files}")
```

```python
# DataFrames
func_dfs = [pd.read_csv(root_path+file) for file in function_files]
invoc_dfs = [pd.read_csv(root_path+file) for file in invocation_files]
app_dfs = [pd.read_csv(root_path+file) for file in app_files]

func_df = pd.concat(func_dfs)
invoc_df = pd.concat(invoc_dfs)
app_df = pd.concat(app_dfs)

print(f"Functions: {len(func_df)}")
print(f"Invocations: {len(invoc_df)}")
print(f"Apps: {len(app_df)}")
```
